Data Visualization

Task: Plot a histogram of the Sentiment column to visualize its distribution.
Question: Does the Sentiment data appear to be normally distributed, or does it lean toward positive/negative sentiments?

```python
import matplotlib.pyplot as plt

# Histogram of Sentiment
plt.figure(figsize=(8, 6))
plt.hist(df['Sentiment'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Frequency')
plt.show()
```

Task: Plot a scatter plot of Likes vs. Retweets to see if there is any correlation between the two.
Question: Is there a correlation between the number of Likes and Retweets?

```python
# Scatter plot of Likes vs Retweets
plt.figure(figsize=(8, 6))
plt.scatter(df['Likes'], df['Retweets'], alpha=0.5, color='purple')
plt.title('Likes vs Retweets')
plt.xlabel('Likes')
plt.ylabel('Retweets')
plt.show()
```

Task: Create a bar plot to visualize the average Sentiment by Day of Week.
Question: Which day of the week has the highest average Sentiment?

```python
# Average Sentiment by Day of Week
avg_sentiment_by_day = df.groupby('Day_of_Week')['Sentiment'].mean()

# Bar plot of average Sentiment by Day of Week
avg_sentiment_by_day = avg_sentiment_by_day.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.figure(figsize=(10, 6))
avg_sentiment_by_day.plot(kind='bar', color='lightcoral')
plt.title('Average Sentiment by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Average Sentiment')
plt.xticks(rotation=45)
plt.show()
```

Text Analysis

Task: Perform a word count analysis on the tweets to see which words are most frequently used.
Question: What are the most common words used in the tweets? Can you identify any significant patterns in the vocabulary of specific categories (e.g., fitness, tech)?

```python
from collections import Counter
import re

# Tokenize the tweets and clean the text
all_words = ' '.join(df['Tweet']).lower()
all_words = re.sub(r'[^\w\s]', '', all_words)

# Count the frequency of words
word_counts = Counter(all_words.split())

# Display the 10 most common words
print(word_counts.most_common(10))
```

Time Series Analysis

Task: Group the data by hour and calculate the total number of Likes and Retweets for each hour of the day.
Question: What time of day (hour) has the highest total engagement (Likes + Retweets)?

```python
# Group by hour and calculate total Likes and Retweets
hourly_engagement = df.groupby('Hour')[['Likes', 'Retweets']].sum()

# Display hourly engagement
print(hourly_engagement)
```
